# 02_Healthcare Costs Analysis

```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt

# Load the dataset
df=pd.read_csv('healthcare_dataset.csv')

# Data overview
print(df.head())
print(df.info())
print(df.describe())

# Data cleaning
df=df.drop_duplicates()
print(df.isnull().sum())
df = df.dropna()

## Average Billing Amount by Medical Condition

plt.figure(figsize=(14, 8))
sbn.barplot(x='Medical Condition', y='Billing Amount', data=df, estimator=sum)
plt.title('Total Billing Amount by Medical Condition')
plt.xlabel('Medical Condition')
plt.ylabel('Total Billing Amount')
plt.xticks(rotation=45)
plt.show()

## Average Billing Amount by Insurance Provider

plt.figure(figsize=(14, 8))
sbn.boxplot(x='Insurance Provider', y='Billing Amount', data=df)
plt.title('Total Billing Amount by Insurance Provider')
plt.xlabel('Insurance Provider')
plt.ylabel('Total Billing Amount')
plt.xticks(rotation=45)
plt.show()

## Total Billing Amount by Insurance Provider

billing_by_insurance = df.groupby('Insurance Provider')['Billing Amount'].sum().reset_index()
plt.figure(figsize=(14, 8))
plt.pie(billing_by_insurance['Billing Amount'], labels=billing_by_insurance['Insurance Provider'], autopct='%1.1f%%', startangle=140)
plt.title('Total Billing Amount by Insurance Provider')
plt.axis('equal') 
plt.show()

## Billing Amount vs. Age by Admission Type

plt.figure(figsize=(14, 8))
sbn.scatterplot(x='Age', y='Billing Amount', hue='Admission Type', data=df)
plt.title('Billing Amount vs. Age by Admission Type')
plt.xlabel('Age')
plt.ylabel('Billing Amount')
plt.show()
```
