# Patient Demographics Analysis

```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt

# Load the dataset
df=pd.read_csv('healthcare_dataset.csv')

# Data overview
print(df.head())
print(df.info())
print(df.describe())

# Data cleaning
df=df.drop_duplicates()
print(df.isnull().sum())
df = df.dropna()

# Age distribution
plt.figure(figsize=(10, 6))
sbn.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Gender distribution
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%01.2f%%', startangle=140)
plt.title('Gender Distribution of Patients')
plt.show()

plt.figure(figsize=(10, 6))
sbn.countplot(x='Gender', data=df)
plt.title('Gender Distribution of Patients')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Blood type distribution
plt.figure(figsize=(10, 6))
sbn.countplot(x='Blood Type', data=df, order=df['Blood Type'].value_counts().index)
plt.title('Blood Type Distribution of Patients')
plt.xlabel('Blood Type')
plt.ylabel('Count')
plt.show()

# Summary
age_summary = df['Age'].describe()
print("Age Summary:\n", age_summary)

gender_summary = df['Gender'].value_counts(normalize=True) * 100
print("\nGender Summary (in %):\n", gender_summary)

blood_type_summary = df['Blood Type'].value_counts(normalize=True) * 100
print("\nBlood Type Summary (in %):\n", blood_type_summary)
```
